```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from arch import arch_model
from scipy.stats import skew, kurtosis
from tqdm import tqdm

def simulate_arch_data(T, omega, alpha):
    x = np.zeros(T)
    sigma2 = np.zeros(T)
    sigma2[0] = omega / (1 - alpha) 
    z=np.random.normal(size=T)
    x[0] = np.sqrt(sigma2[0]) *z[0]
    
    for t in range(1, T):
        sigma2[t] = omega + alpha * x[t-1]**2
        x[t] = np.sqrt(sigma2[t]) * z[t]
    
    return x

n= 10000
T_values = [250,500,1000, 2000, 5000, 10000, 20000]
omega = 1 
alpha = 0.5


for T in T_values:
    alpha_hats = []
    omega_hats = []
    mu_hats=[]

    for _ in  tqdm(range(n), desc=f"T = {T}"):
        x = simulate_arch_data(T, omega, alpha)
        model = arch_model(x, vol='ARCH', p=1, mean='constant', rescale=False)
        res = model.fit(disp='off')
        alpha_hats.append(res.params['alpha[1]'])
        omega_hats.append(res.params['omega'])
        mu_hats.append(res.params['mu'])
        

    for param, values, true_val, label in zip(
        ['Alpha', 'Omega', 'Mu'], 
        [alpha_hats, omega_hats, mu_hats], 
        [alpha, omega, 0], 
        ['Alpha', 'Omega', 'Mu']
    ):
        plt.figure(figsize=(10, 5))
        sns.histplot(values, bins=50, kde=True, color='skyblue', edgecolor='k')
        plt.axvline(x=true_val, color='red', linestyle='--', label=f'True {param} = {true_val}')
        plt.title(f"Sampling Distribution of {label} (T = {T}, {n} simulations)")
        plt.xlabel(f'{label} Values')
        plt.ylabel('Frequency')
        plt.legend()
        plt.tight_layout()
        plt.show()

        print(f"\nMonte Carlo Summary for T = {T}, parameter {param}:")
        print(f"  Mean({param}̂)   = {np.mean(values):.4f}")
        print(f"  Bias({param}̂)   = {np.mean(values) - true_val:.4f}")
        print(f"  Std Dev          = {np.std(values):.4f}")
        print(f"  Skewness         = {skew(values):.4f}")
        print(f"  Kurtosis         = {kurtosis(values, fisher=False):.4f}")
        print("-" * 40)
```
